```python
import pandas as pd

scores = [
    {'chinese':78,'english':80,'math':92,'history':75,'discover':85},
    {'chinese':84,'english':90,'math':98,'history':85,'discover':95}

]

dataFrame = pd.DataFrame(index=['robert','jenny'],data=scores)
display(dataFrame)
```

```python
subjects = list(dataFrame.columns)
subjects
```

```python
dataFrame['chinese']
```

```python
#T,index和column對調
swapDataFrame = dataFrame.T
display(swapDataFrame)
robert_series = swapDataFrame['robert']
display(robert_series)
jenny_series = swapDataFrame['jenny']
jenny_series
```

```python
robert_scores = robert_series.to_list()
jenny_scores = jenny_series.to_list()
display(subjects)
display(robert_scores)
display(jenny_scores)
```

```python
import matplotlib.pyplot as plt
figure = plt.figure()
axes = figure.add_subplot(1,1,1)
axes.plot(subjects,robert_scores,'bo-')
axes.plot(subjects,jenny_scores,'ro-')
plt.show()
```
