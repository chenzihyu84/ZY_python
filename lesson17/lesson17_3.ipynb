```python
import pandas as pd
#選取需要的欄和列
dataFrame1 = pd.read_csv('各鄉鎮市區人口密度.csv')
dataFrame1.columns = ['統計年','區域別','年底人口數','土地面積','人口密度']
dataFrame2 = dataFrame1.drop(index=0)
dataFrame3 = dataFrame2.reindex(index=range(1,371))
dataFrame3
```

```python
#value的檢查,有沒有空的欄位,資料類型
dataFrame3.info()
```

```python
def toInt(value:str) -> int:
    try:
        newValue = int(value)
    except:
        newValue = 0
    return newValue

def toFloat(value:str) -> float:
    try:
        newValue = float(value)
    except:
        newValue = 0.0
    return newValue

dataFrame3['年底人口數'] = dataFrame3['年底人口數'].map(toInt)
dataFrame3['人口密度'] = dataFrame3['人口密度'].map(toInt)
dataFrame3['土地面積'] = dataFrame3['土地面積'].map(toFloat)
dataFrame3
```

```python
#subscript[slice]
dataFrame4 = dataFrame3.iloc[:5]
#使用比較運算子
dataFrame4['年底人口數'] > 400000
#使用subscript[[bool]]
dataFrame4[dataFrame4['年底人口數'] > 400000]
```
