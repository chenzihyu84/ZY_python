```python
import numpy as np
import pandas as pd
score = np.random.randint(50,101,size=(50,5))
score_pd = pd.DataFrame(score,columns=['國文','英文','數學','地理','歷史'],index=range(1,51))
score_pd
```

```python
score_pd.index.name="學號"
score_pd.columns.name="科目"
score_pd
```

```python
score_pd['總分'] = score_pd[['國文','英文','數學','地理','歷史']].sum(axis=1)
score_pd
score_pd['平均'] = score_pd[['國文','英文','數學','地理','歷史']].mean(axis=1)
score_pd
```

```python
score_pd['排名'] = score_pd['總分'].rank(method='min',ascending=False)
score_pd
```

```python
score_pd.sort_values(by='排名')
```
